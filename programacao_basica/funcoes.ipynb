<h1> Funções </h1>

Para definir funções em python, é necessário utilizar a palavra 'def'.

Segue exeplo de uma função capaz de somar dois números.
```
# Função com parâmetros (a e b) e com retorno.
def calculateSum(a, b):
	return (a+b)
```

Também conhecidas como sub-rotinas, as funções são muito utilizadas em programação.

Um dos grandes benefícios é não precisar ficar copiando o código toda vez que executar uma determinada operação.

As funções serão um dos nossos principais recursos quando construirmos quantidades cada vez maiores de código para resolver problemas.

A função, por definição, é um dispositivo que agrupa um conjunto de instruções para que elas possam ser executadas mais de uma vez. Funções também permitem especificar os parâmetros que podem servir como entrada para as funções.

Em um nível mais fundamental, a construção de funções nos permite reutilizar código, sem ter que escrevê-lo novamente. Nas aulas de Strings, utilizamos a função len() para obter o comprimento de uma string. Com funções, escrevemos o código uma única vez e repetimos a mesma instrução, fazendo a chamada à função, quantas vezes forem necessárias.

O formato geral de uma função é:
```
def nome_da_funcao(arg1, arg2, etc...):
	"Aqui podem vir comentários, documentando sua função"
	
	<Aqui vai seu código>
	<Retorno desejado pela função> 

```

E por que função é importante?

Funções em Python são uma forma de escrever a sua lógica em único pacote e utilizá-la em diferentes lugares no seu código e quantas vezes quiser.

Uma função pode conter variáveis locais, que são conhecidas pelo código geral somente dentro da função, e demais variáveis devem ser passadas como argumento (também chamado de parâmetro).

Definindo uma função:

````python
def primeiraFunc(): 
  print('Hello World')
````

Definindo uma função com parâmetro:

````python
def primeiraFunc(nome): 
  print('Hello %s' %(nome))
````

````python
primeiraFunc('Aluno')
````

Um for dentro de uma função:

````python
def funcLeitura():
  for i in range(0, 5):
    print('Número ' + str(i))
````

````python
funcLeitura()
````

Função para somar número:

````python
def addNum(n1, n2):
  soma = n1 + n2
  print('Soma: ', soma)
````

````python
addNum(2, 9)
````

Fazendo uso de variável global:

````python
var_global = 10  # Variavel global

def multiplica(n1, n2):
  var_global = n1 * n2
  print(var_global) # Variaval local
````

````python
multiplica(20, 20)
````

````python
print(var_global)
````

Executando as 3 células acima podemos notar que uma variável 'var_global' é mencionada duas vezes no código. Porém, mesmo com elas sendo escritas da mesma maneira elas não valem a mesma coisa.

A variável global pode ser mencionada em qualquer parte do código, já a local (que está dentro da função) não. Veja novamente:

````python
var_global = 10 

def multiplica(n1, n2):
  var_local = n1 * n2
  print(var_local)

print(var_global)
print(var_local)
````

O erro se dá justamente pela 'não existência' de 'var_local' fora do escopo da função.

Você também pode criar funções usando outras funções.

````python
import math

def numPrimo(num):
  # Verificando se um número é primo
  if num % 2 == 0 and num > 2:
    return 'Este número não é primo'
  for i in range(3, int(math.sqrt(num) + 1), 2):
    if num % i == 0:
      return 'Este número não é primo'
  return 'Este número é primo'
````

````python
numPrimo(541)
````

Note o uso da função sqrt() que está inserida no pacote math.

Pacotes são conjuntos de diversas funções. O pacote math é um pacote focado em funções de uso matemático, neste caso, achando a raiz quadrada do número recebido como argumento.

Fazendo split (separação) de dados:

````python
def split_string(text):
  return text.split(" ")
````

````python
text = 'Esta função será bastante útil para separar grandes volumes de dados'

new_text = split_string(text)

print(new_text)
````

O **Split** é usado para dividir uma string em pequenos pedaços. Para isso, seu método retorna um array de strings, contendo as respectivas partes, definidas de acordo com a string passada como parâmetro, que na verdade funciona como o “agente” divisor da mesma.

Em Phyton, a função split pode receber um argumento e assim fazer a divisão a partir daquele caractere.

<h1> Número de argumentos variável em funções </h1>
Python possui uma sintaxe que permite que definamos uma função com um número indefinido de argumentos. Isso pode ser feito através do *(asterisco). Por exemplo:

```
def func(x, *args):
    print x
    print args

func(1, 2, 3, 4, 5)
```

Execute o código acima, e verá a seguinte saída:

```
1
(2, 3, 4, 5)
```
Ou seja, o primeiro argumento fornecido pelo usuário (o inteiro 1) é atribuído ao primeiro parâmetro da função. Os argumentos restantes (os inteiros 2, 3, 4 e 5) são empacotados em uma tupla e passados ao parâmetro ***args**. 

Obs.: Mais tarde veremos o que são tuplas, não se preocupe.

Desse modo, os elementos de args podem ser acessados como os elementos de uma tupla qualquer, seja por índice, ou realizando uma travessia com **for in**. Por exemplo, uma função que realiza a soma de todos os argumentos recebidos:

````python
def soma(*args):
    result = 0
    for elem in args:
        result += elem
    return result

soma(1, 2, 3, 4, 5)
````

Também podemos usar o '*' em frente ao nome de sequências (como tuplas, listas,) para fazer com que os valores que compõem tais sequências sejam desmembrados e passados como argumentos para funções. Considere e função func() abaixo:

```
def func(a,b,c,d):
    print a,b
```

Ela possui 4 parâmetros (aqui representados pelas variáveis a, b, c, d) e imprime os valores de a
 e de b. 

Imagine agora que você possui uma lista com 4 valores, que deseja passar para a função func(). Poderia fazer o seguinte:

```
l = [1, 2, 3, 4]
func(l[0], l[1], l[2], l[3])
```
Correto? Sim. Mas dá pra fazer melhor. E se eu quisesse passar a lista toda para a função, sem precisar acessar os elementos individuais.
```
l = [1, 2, 3, 4]
func(l)
```
**Errado!**
 Afinal, a função espera 4 valores e estamos passando somente um, uma lista. É aí que podemos usar o *.
```
l = [1, 2, 3, 4]
func(*l)
```
